```python
# Gabungkan DataFrames transaction_df dengan store_df berdasarkan kolom "StoreID"
combined_df = transaction_df.merge(store_df, on="StoreID", how="left")

# Gabungkan DataFrames combined_df dengan product_df berdasarkan kolom "ProductID"
combined_df = combined_df.merge(product_df, on="ProductID", how="left")

# Gabungkan DataFrames combined_df dengan customer_df berdasarkan kolom "CustomerID"
combined_df = combined_df.merge(customer_df, on="CustomerID", how="left")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
# Pilih variabel target (y) dan fitur (X)
X = combined_df[['Month', 'Day', 'Price_x', 'Qty', 'Price_y']]
y = combined_df['TotalAmount']
```

```python
# Bagi data menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)  # shuffle=False untuk data time series
```

```python
# Membuat model regresi (Linear Regression)
model = LinearRegression()
```

```python
# Melatih model pada data latih
model.fit(X_train, y_train)
```

```python
# Melakukan prediksi pada data uji
y_pred = model.predict(X_test)
```

```python
# Evaluasi model
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f'MAE: {mae}')
print(f'MSE: {mse}')
print(f'RMSE: {rmse}')
print(f'R-squared (R2): {r2}')
```

```python
# Visualisasi hasil
plt.figure(figsize=(12, 6))
plt.plot(y_test.index, y_test, label='Actual TotalAmount', marker='o')
plt.plot(y_test.index, y_pred, label='Predicted TotalAmount', marker='x')
plt.xlabel('Date')
plt.ylabel('TotalAmount')
plt.legend()
plt.title('Actual vs. Predicted TotalAmount')
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# Pilih fitur yang akan digunakan untuk clustering
# Dalam contoh ini, saya menggunakan 'Price_x' dan 'Price_y' sebagai fitur
X = combined_df[['Price_x', 'Price_y']]
```

```python
# Menentukan jumlah cluster (silhouette analysis dapat digunakan untuk menentukan jumlah cluster yang optimal)
n_clusters = 3
```

```python
# Membuat model K-Means
kmeans = KMeans(n_clusters=n_clusters, random_state=0)
```

```python
# Melakukan clustering
combined_df['Cluster'] = kmeans.fit_predict(X)
```

```python
# Visualisasi hasil clustering
plt.scatter(X['Price_x'], X['Price_y'], c=combined_df['Cluster'], cmap='rainbow')
plt.xlabel('Price_x')
plt.ylabel('Price_y')
plt.title(f'K-Means Clustering with {n_clusters} Clusters')
plt.show()
```
